```python
import pandas as pd
import os
```

```python
original_df = pd.read_csv("../exported/news.csv")
original_df
```

```python
import json

dates = []
for news_id in list(original_df['id']):
    with open(f"../fakenewsnet_dataset/politifact/fake/{news_id}/first_tweet_date.json", 'r') as f:
        date_obj = json.loads(f.read())
        dates.append(f"{date_obj['year']}-{date_obj['month']}-{date_obj['day']}")
        
dates
```

```python
new_df = original_df.copy()
```

```python
new_df['date'] = dates
new_df
```

```python
import csv
new_df.to_csv("../exported/news_with_date.csv", header = True, index = False, quoting = csv.QUOTE_NONNUMERIC, quotechar='"')
```
